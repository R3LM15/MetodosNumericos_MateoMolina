### 3. Dir√≠jase al pseudoc√≥digo del spline c√∫bico con frontera natural provisto en clase, en base a ese pseudoc√≥digo complete la siguiente funci√≥n:

```python
import sympy as sym
from IPython.display import display


# ###################################################################
def cubic_spline(xs: list[float], ys: list[float]) -> list[sym.Symbol]:

    points = sorted(zip(xs, ys), key=lambda x: x[0])  # sort points by x

    xs = [x for x, _ in points]
    ys = [y for _, y in points]

    n = len(points) - 1  # number of splines

    h = [xs[i + 1] - xs[i] for i in range(n)]  # distances between  contiguous xs

    alpha = [0] * n
    for i in range(1, n):
        alpha[i] = 3 / h[i] * (ys[i + 1] - ys[i]) - 3 / h[i - 1] * (ys[i] - ys[i - 1])

    l = [1]
    u = [0]
    z = [0]

    for i in range(1, n):
        l += [2 * (xs[i + 1] - xs[i - 1]) - h[i - 1] * u[i - 1]]
        u += [h[i] / l[i]]
        z += [(alpha[i] - h[i - 1] * z[i - 1]) / l[i]]

    l.append(1)
    z.append(0)
    c = [0] * (n + 1)

    x = sym.Symbol("x")
    splines = []
    for j in range(n - 1, -1, -1):
        c[j] = z[j] - u[j] * c[j + 1]
        b = (ys[j + 1] - ys[j]) / h[j] - h[j] * (c[j + 1] + 2 * c[j]) / 3
        d = (c[j + 1] - c[j]) / (3 * h[j])
        a = ys[j]
        print(j, a, b, c[j], d)
        S = a + b * (x - xs[j]) + c[j] * (x - xs[j]) ** 2 + d * (x - xs[j]) ** 3

        splines.append(S)
    splines.reverse()
    return splines
```

### 4. Usando la funci√≥n anterior, encuentre el spline c√∫bico para:


```python
xs = [1, 2, 3]
ys = [2, 3, 5]

splines = cubic_spline(xs=xs, ys=ys)
_ = [display(s) for s in splines]
print("______")
_ = [display(s.expand()) for s in splines]
```

### 6. Use la funci√≥n cubic_spline_clamped, provista en el enlace de Github, para graficar los datos de la siguiente tabla.



```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt


def cubic_spline_clamped(xs, ys, B0, B1):
    """
    Calcula y grafica splines c√∫bicos con condiciones de derivada en los extremos (clamped).
    """

    # Ordenar puntos por x
    points = sorted(zip(xs, ys), key=lambda x: x[0])
    xs = [x for x, _ in points]
    ys = [y for _, y in points]
    n = len(xs) - 1
    h = [xs[i + 1] - xs[i] for i in range(n)]

    # Paso 1: calcular alpha
    alpha = [0] * (n + 1)
    alpha[0] = 3 / h[0] * (ys[1] - ys[0]) - 3 * B0
    alpha[-1] = 3 * B1 - 3 / h[-1] * (ys[-1] - ys[-2])
    for i in range(1, n):
        alpha[i] = 3 / h[i] * (ys[i + 1] - ys[i]) - 3 / h[i - 1] * (ys[i] - ys[i - 1])

    # Paso 2: resolver sistema tridiagonal
    l = [2 * h[0]]
    u = [0.5]
    z = [alpha[0] / l[0]]
    for i in range(1, n):
        l.append(2 * (xs[i + 1] - xs[i - 1]) - h[i - 1] * u[i - 1])
        u.append(h[i] / l[i])
        z.append((alpha[i] - h[i - 1] * z[i - 1]) / l[i])
    l.append(h[-1] * (2 - u[-1]))
    z.append((alpha[-1] - h[-1] * z[-1]) / l[-1])

    # Paso 3: construir coeficientes
    c = [0] * (n + 1)
    c[-1] = z[-1]
    for j in range(n - 1, -1, -1):
        c[j] = z[j] - u[j] * c[j + 1]

    x_sym = sym.Symbol("x")
    splines = []
    for j in range(n):
        b = (ys[j + 1] - ys[j]) / h[j] - h[j] * (c[j + 1] + 2 * c[j]) / 3
        d = (c[j + 1] - c[j]) / (3 * h[j])
        a = ys[j]
        S = a + b * (x_sym - xs[j]) + c[j] * (x_sym - xs[j]) ** 2 + d * (x_sym - xs[j]) ** 3
        splines.append((S, (xs[j], xs[j + 1])))

    # Gr√°fica
    plt.figure(figsize=(10, 5))
    for S, (xi, xf) in splines:
        f_lambd = sym.lambdify(x_sym, S, modules=["numpy"])
        x_vals = np.linspace(xi, xf, 100)
        y_vals = f_lambd(x_vals)
        plt.plot(x_vals, y_vals, label=f"${sym.simplify(S)}$")

    # Graficar puntos originales
    plt.plot(xs, ys, "o", color="black", label="Datos")

    plt.title("Spline C√∫bico con Condiciones de Borde (Clamped)")
    plt.xlabel("x")
    plt.ylabel("S(x)")
    plt.grid(True)
    plt.legend(loc="best")
    plt.tight_layout()
    plt.show()

    return [S for S, _ in splines]  # devuelve solo las expresiones


```
