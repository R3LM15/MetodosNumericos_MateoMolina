1. Use el método de bisección para encontrar soluciones precisas dentro de  $ 10^{-2}$ para $ x^{3} -7x^{2} + 14x-6=0$  en cada intervalo.

    a. $ [0,1] $
    b. $[1, 3.2]$
    c. $[3.2, 4]$ 

```python

def f(x):
    return x**3 - 7*x**2 + 14*x - 6  

def biseccion(a, b, tol=1e-2, max_iter=50):
    print(f"\nIntervalo inicial: [{a}, {b}]")
    if f(a) * f(b) >= 0:
        print("El método de bisección no se puede aplicar: f(a) y f(b) deben tener signos opuestos.")
        return None

    iteracion = 0
    while abs(b - a) > tol and iteracion < max_iter:
        c = (a + b) / 2.0

        if f(c) == 0:
            return c  # Se encontró la raíz exacta
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iteracion += 1




        print(f"Iteración {iteracion}: c = {c}, f(c) = {f(c)} rango = [{a}, {b}]")
    return (a + b) / 2.0

# Intervalos
raiz = biseccion(0, 1)
print(f"\nRaíz aproximada: {raiz}")

# Intervalos
raiz = biseccion(1, 3.2)
print(f"\nRaíz aproximada: {raiz}")

# Intervalos
raiz = biseccion(3.2, 4)
print(f"\nRaíz aproximada: {raiz}")
```

```python

def f(x):
    return (x+3)*(x+1)**2 *x*(x-1**3)**3 *(x-3) 

def biseccion(a, b, tol=0, max_iter=100):
    print(f"\nIntervalo inicial: [{a}, {b}]")
    if f(a) * f(b) >= 0:
        print(f"El método de bisección no se puede aplicar: f(a) = {f(a)} y f(b) = {f(b)}  deben tener signos opuestos.")
        return None

    iteracion = 0
    while abs(b - a) > tol and iteracion < max_iter:
        c = (a + b) / 2.0

        if f(c) == 0:
            return c  # Se encontró la raíz exacta
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iteracion += 1

        print(f"Iteración {iteracion}: c = {c}, f(c) = {f(c)} rango = [{a}, {b}]")
    return (a + b) / 2.0

# Intervalos
raiz = biseccion(-1.5,2.5)
print(f"\nRaíz aproximada: {raiz}")

# Intervalos
raiz = biseccion(-0.5, 2.4)
print(f"\nRaíz aproximada: {raiz}")

# Intervalos
raiz = biseccion(-0.5, 3)
print(f"\nRaíz aproximada: {raiz}")

#intervalo de prueba
raiz = biseccion(-0.5, 3.5)
print(f"\nRaíz aproximada: {raiz}")
```

$$ S(t) = S_0 - \frac{mg}{k} t + \frac{m^{2}g}{k^{2}} ( 1 - e^{-kt/m} ) $$ 

donde g = 9.81 $m/s^{2}$ y k representa el coeficiente de la resistencia del aire en $Ns/m$. Suponga $S_0$ = 300m, m = 0.25 kg y k = 0.1 Ns/m. Encuentre dentro de 0.01 𝑠𝑒𝑔𝑢𝑛𝑑𝑜𝑠, el tiempo que tarda un cuarto de kg en golpear el piso.
